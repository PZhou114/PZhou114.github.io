Exercise: Newton-Raphson Method (see the beamer presentation)

Given that $f(x) = x^3 + 2x - 2$, then $f'(x) = 3x^2 + 2$. The Newton-Raphson recursion would be in the form of
$$
x_{n+1} = x_n - \frac{x_n^3 + 2x_n -2}{3x_n^2 + 2}
$$

A reasonable initial guess might be the 0.

```python
# Exercise: Newton-Raphson method (using for loop)

x = 0  # initial guess
x_recursion_list = []  # construct a list storing all iterative results

for i in range(10):  # we run 10 iterations (10 is enough as the convergence is rather fast)
    x_new = x - (x**3 + 2*x - 2)/(3 * x**2 + 2)   # Newton-Raphson recursion
    x_recursion_list.append(x_new)  # append 
    x = x_recursion_list[i]  # update x

import matplotlib.pyplot as plt
plt.plot(x_recursion_list, marker = '.')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title('Newton-Raphson Method by using for loop')
plt.grid()
plt.show()
```

If applying the <code>while</code> loop, a commonly used stopping criteria would be the difference between the last value and the current value: we would stop the iteration if the difference is less than a certain threshold, which is always set to be very small to yield the convergence behaviour. 

```python
# Exercise: Newton-Raphson method (using while loop)

x = 2  # initial guess
x_recursion_list = []  # construct a list storing all iterative results

eps = 1e-6  # stopping threshold
diff = 1  # initialise the recursive difference

while diff > eps:
    x_new = x - (x**3 + 2*x - 2)/(3 * x**2 + 2)   # Newton-Raphson recursion
    x_recursion_list.append(x_new)  # append 
    diff = np.abs(x_new - x)  # absolute difference
    x = x_new  # update x
    
plt.plot(x_recursion_list, marker = '.')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title('Newton-Raphson Method by using while loop')
plt.grid()
plt.show()
```

```python
def F1_WDC(year):
    """
    Check the WDC of F1 for the most recent 10 years.
    """

    if year < 2015:
        print('Our searching year should be no earlier than 2015.')
    elif year in set({2015, 2017, 2018, 2019, 2020}):
        print(f'F1 WDC in {year} is Lewis Hamilton.')
    elif year in set({2021, 2022, 2023, 2024}):
        print(f'F1 WDC in {year} is Max Verstappen.')
    else:
        print(f'F1 WDC in {year} is Nico Rosberg.')

F1_WDC(2016)
```

```python
# Exercise: Newton-Raphson Method

def Newton_Raphson(f, df, init_guess=0, iter_num=10):
    x = init_guess  # initial guess
    x_recursion_list = []  # construct a list storing all iterative results

    for i in range(iter_num):  # we run 10 iterations (10 is enough as the convergence is rather fast)
        x_new = x - (f(x))/(df(x))   # Newton-Raphson recursion
        x_recursion_list.append(x_new)  # append 
        x = x_recursion_list[i]  # update x

    return x_recursion_list
```

```python
import sympy as sp

x = sp.symbols('x')
expr = x**3 + 2*x - 2
dexpr = sp.diff(expr, x)

f = sp.lambdify(x, expr, 'numpy')
df = sp. lambdify(x, dexpr, 'numpy')

Newton_Raphson(f, df)
```
